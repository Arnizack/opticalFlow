# Launch Parameters

## TilesBuffer

Sei:
\begin{equation}
D_I = \text{Dimension des Image} \\
D_G  = \text{Dimension des Grids} \\
T = \text{Tiles Size} \\
P = \text{Padding Size} \\
S = \text{Shared Memory} \\
\end{equation}

Dann ist die Größe des Shared Memory:
    \begin{equation}
    S=\left( \left\lceil \frac{D_{I,1}}{T_1 D_{G,1}} \right\rceil T_1 + 2 P_1 \right)  \left( \left\lceil \frac{D_{I,2}}{T_2 D_{G,2}} \right\rceil T_2 + 2 P_2 \right)
\leq \left( \frac{D_{I,1}}{D_{G,1}} + T_1 + 2 P_1 \right)  \left( \frac{D_{I,2}}{D_{G,2}} + T_2 + 2 P_2 \right) \\
= \frac{D_{I,1}D_{I,2}}{D_{G,1}D_{G,2}} 
+ \left(T_2 + 2P_2 \right) \frac{D_{I,1}}{D_{G,1}}
+ \left(T_1 + 2P_1 \right)  \frac{D_{I,2}}{D_{G,2}}
+ \left(T_1 + 2P_1 \right)\left(T_2 + 2P_2 \right) \\
\leq S_{\text{max}}
\end{equation}


Minimiere den speicher aller Blöcke zusammen, um die lade Zeit zu minimieren.

\begin{equation}
S_\text{ges}(D_{G,1},D_{G,1}) = 
D_{G,1} D_{G,2} \left( \frac{D_{I,1}}{D_{G,1}} + T_1 + 2 P_1 \right)  \left( \frac{D_{I,2}}{D_{G,2}} + T_2 + 2 P_2 \right) \\
= D_{I,1}D_{I,2}
+ \left(T_2 + 2P_2 \right) D_{G,2} D_{I,1}
+ \left(T_1 + 2P_1 \right) D_{G,1} D_{I,2}
+ D_{G,1} D_{G,2} \left(T_1 + 2P_1 \right)\left(T_2 + 2P_2 \right)
\end{equation}

Gesucht ist $ D_G $.
Somit setze $ x:=D_G $.

\begin{equation}
= D_{I,1}D_{I,2}
+ \left(T_2 + 2P_2 \right) D_{I,1}  x_2
+ \left(T_1 + 2P_1 \right) D_{I,2} x_1 
+ \left(T_1 + 2P_1 \right)\left(T_2 + 2P_2 \right) x_1 x_2 
\end{equation}

Für mehrere Tiles sehen die Bedingungen so aus:

Sei:

\begin{equation}
b_0 = \sum_i D_{I,1,i}D_{I,2,i} \\
b_1 = \sum_i  D_{I,2,i} \left(T_{1,i} + 2P_{1,i} \right) \\
b_2 = \sum_i  D_{I,1,i} \left(T_{2,i} + 2P_{2,i} \right) \\
b_3 = \sum_i  \left(T_{1,i} + 2P_{1,i} \right) \left(T_{2,i} + 2P_{2,i} \right) \\
\end{equation}

Dann ist die Limitierung: 
\begin{equation}
\frac{b_0}{x_1 x_2} 
+ \frac{b_2}{x_1}
+ \frac{b_1}{x_2}
+ b_3 
\leq S_{\text{max}} \\
\Leftrightarrow \\
b_0 
+ b_2 x_2 
+ b_1 x_1 
+ b_3 x_1 x_2
\leq S_{\text{max}} x_1 x_2 \\
\end{equation}

und

\begin{equation}
S_\text{ges}(x_1,x_2) = b_0 + b_1 x_1 + b_2 x_2 + b_3 x_1 x_2
\end{equation}

$S_{\text{ges}}$ monoton steigend in $x_1 \geq 0$ $x_2\geq0$ 

$\Rightarrow$
Minimum bei $x_1 = 0 , x_2= 0$

Da die Bedingung $x_1$ und $x_2$ von dem Ursprung abhält, liegt das Minimum auf 
\begin{equation}
b_0 
+ b_2 x_2 
+ b_1 x_1 
+ b_3 x_1 x_2
= S_{\text{max}} x_1 x_2
\end{equation}

## Lagrange Multiplikator

Bedingung:

$$ g(x_1,x_2) = b_0 + b_2 x_2 + b_1 x_1 + b_3 x_1 x_2 - S_{\text{max}} x_1 x_2 $$

Es muss gelten:

\begin{align}
\nabla_{x_1} g = b_1 + ( b_3-S_{\text{max}} ) x_2 \neq 0\\
\nabla_{x_2} g = b_2 + ( b_3-S_{\text{max}} ) x_1 \neq 0\\
\end{align}

Lagrange Funktion:

$$ \Lambda(x_1,x_2) = b_0 + b_1 x_1 + b_2 x_2 + b_3 x_1 x_2 + \lambda g(x_1,x_2) $$

\begin{align}
\nabla_{x_1,x_2,\lambda} \Lambda  =
\left\{
\begin{array}
    \text{ } b_1 + b_3 x_2 + \lambda b_1 + \lambda b_3 x_2 - \lambda S_{\text{max}} x_2 \\
    b_2 + b_3 x_1 + \lambda b_2 + \lambda b_3 x_1 - \lambda S_{\text{max}} x_1 \\
    g(x_1,x_2)
\end{array}
\right.
\end{align}


Löse:

$$\nabla_{x_1,x_2,\lambda} \Lambda = 0$$



\begin{equation}
\Rightarrow \\
x_{\text{ext},1,1} = \frac{ b_1 b_2 - \sqrt{b_1 b_2 (b_0 (S - b_3) + b_1 b_2)}}{b_1 (S - b_3)} \\
\wedge \\
x_{\text{ext},1,2} = \frac{ b_1 b_2 - \sqrt{b_1 b_2 (b_0 (S - b_3) + b_1 b_2)}}{b_2 (S - b_3)}\\ 
\vee \\
x_{\text{ext},2,1} = \frac{ b_1 b_2 + \sqrt{b_1 b_2 (b_0 (S - b_3) + b_1 b_2)}}{b_1 (S - b_3)} \\
\wedge \\
x_{\text{ext},2,2} = \frac{ b_1 b_2 + \sqrt{b_1 b_2 (b_0 (S - b_3) + b_1 b_2)}}{b_2 (S - b_3)}
\end{equation}

Bedingung:

$$
b_1 S (S - b_3)\neq0 
\wedge
b_2 \neq0
\wedge
\sqrt{b_1 b_2 (b_0 (S - b_3) + b_1 b_2)}\neq0
$$

$$\Leftarrow b_1 > 0 \wedge b_1 > 0 \wedge b_2 > 0 \wedge b_3 > 0 \wedge S > b_3$$

$\nabla g \neq 0$ für $ x_{\text{ext},1} $


$\nabla g \neq 0$ für $ x_{\text{ext},2} $

Da $g(x)$ keinen Kreis bildet sonder eine eher asymptotische Form, wird das positive $x_{\text{ext}} $ als Minimum gewählt.

Somit nehme $x_{\text{ext,2}} $ als Minimum.
